### Cleaning Data

```python
# Fungsi untuk mengonversi angka menjadi nama musim pada kolom season
def find_season(season):
    season_dict = {
        1: 'Winter',
        2: 'Spring',
        3: 'Summer',
        4: 'Fall'
    }
    return season_dict.get(season, "Unknown season")
df['season']= df['season'].apply(find_season)

print(df)
```

```python
# Fungsi untuk mengonversi angka menjadi nama weathersit pada kolom weathersit
def find_weathersit(weathersit):
    weathersit_dict = {
        1: 'Clear, Few clouds, Partly cloudy, Partly cloudy',
        2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
        3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered cloudsr',
        4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'
    }
    return weathersit_dict.get(weathersit, "Unknown weathersit")
df['weathersit']= df['weathersit'].apply(find_weathersit)

print(df)
```

```python
# Fungsi untuk mengonversi angka menjadi huruf holiday non holiday
def find_holiday(holiday):
    holiday_dict = {
        0: 'non holiday',
        1: 'holiday'
    }
    return holiday_dict.get(holiday, "Unknown weathersit")
df['holiday']= df['holiday'].apply(find_holiday)

print(df)
```

```python
# fungsi untuk mengkonversi angka menjadi huruf pada tabel workingday
def find_workingday(workingday):
    workingday_dict = {
        0: 'non workingday',
        1: 'workingday'
    }
    return workingday_dict.get(workingday, "Unknown workingday")
df['workingday']= df['workingday'].apply(find_workingday)

print(df)
```

```python
#menemukan dtaa weekday yang sesuai
df['dteday'] = pd.to_datetime(df['dteday'])
```

```python
#mengubah wekkday menjadi data yang sesuai
df['weekday'] = df['dteday'].dt.day_name()
```

```python
#mengganti bulan angka(integer) menjadi huruf nama bulannya (string)

df['mnth'] = df['dteday'].dt.month_name()
```

```python
#mengubah yr menjadi tahun yang sesuai dengan dat dteday nya
df['yr'] = df['dteday'].dt.year
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

**pertanyaan:**
-  â€œBagaimana performa pengguna jasa sewa sepeda tiap bulannya ?

```python
# tampilkan histogram tiap bulannya
plt.figure(figsize=(10,6))

sns.lineplot(x="dteday", y="cnt", data=df, color='blue')
plt.xlabel("Tanggal")
plt.ylabel("Total penyewa")
plt.title("grafik persewaan sepeda tiap bulannya pada tahun 2011-2012")
plt.tight_layout()
plt.show()
```

terlihat titik tertinggi untuk pernyewa adalah pada bulan 5 hingga 10 tahun 2012


```python
#mengurutkan terbanyak perbulannya
df.sort_values(by=["mnth", "cnt"], ascending=[False, False])
```

benar saja kita lihat pada saat kita urutkan data terbesar tabel nilai terbanyak penyewa adalah 8714 pada bulan oktober

```python
#membuat tabel data bulan
mean_df = df.groupby(by="mnth").agg({
          "mnth": "nunique",
          "cnt": ["mean", "std", "max", "min"]
})
mean_df = mean_df.reset_index()
# memsisahkan menggunakan _
mean_df.columns = ["_".join(col) for col in mean_df.columns.values]
```

```python
#diagram grafik rata rata mean max min pengguna sepeda tiap bulannya
plt.figure(figsize=(10,6))
sns.lineplot(x=mean_df.index, y=mean_df["cnt_mean"], color='blue')
sns.lineplot(x=mean_df.index, y=mean_df["cnt_std"], color='red')
sns.lineplot(x=mean_df.index, y=mean_df["cnt_max"], color='green')
sns.lineplot(x=mean_df.index, y=mean_df["cnt_min"], color='yellow')
plt.xlabel("bulan")
plt.ylabel("Total penyewa")
plt.title("rata grafik persewaan sepeda tiap bulannya pada tahun 2011-2012")
plt.tight_layout()
plt.show()
```

**Insight:**
- analisis menggunakan rata rata pengguna tiap bulannya. pengguna naik turun pada saat bulan ke 4 terjadi penurunnan drastis. namun selanjutnya naik terjadi penurunan lagi namun tidak sedrastis pada bulan 4 dan terus meningkat.
